```python
import numpy
import math
import matplotlib.pyplot
import random
```

```python
data = open("1.txt")
n = int(data.readline())
ntrain = int(data.readline())
xtrain = []
ytrain = []
for i in range(ntrain):
    helpx = list(map(int, data.readline().split()))
    ytrain.append(helpx[-1])
    helpx.pop()
    xtrain.append(helpx)
ntest = int(data.readline())
xtest = []
ytest = []
for i in range(ntest):
    helpx = list(map(int, data.readline().split()))
    ytest.append(helpx[-1])
    helpx.pop()
    xtest.append(helpx)
```

# Метод наименьших квадратов

```python
def firstMethod(x, y):
    svd = list(numpy.linalg.svd(x, full_matrices=False))
    svd[0] = numpy.transpose(svd[0])
    svd[1] = numpy.linalg.inv(numpy.diag(svd[1]))
    svd[2] = numpy.transpose(svd[2])
    fplus = numpy.dot( numpy.dot (svd[2], svd[1]), svd[0])
    return numpy.transpose (numpy.dot(fplus, y))
```

```python
def nrmse (actual, predicted):
    import sys
    numpy.set_printoptions(threshold=sys.maxsize)
    print(numpy.square(numpy.array(predicted) - numpy.array(actual)))
    return numpy.sqrt(numpy.mean(numpy.square(predicted - actual))) / (numpy.max(actual) - numpy.min(actual))
```

```python
def firstMethodError():
    print (nrmse(ytest, numpy.dot(xtest, firstMethod(xtrain, ytrain))))
```

```python
firstMethodError()
```

# Градиентный спуск


```python
def secondMethod(x, y, k, nu):
    w = []
    for i in range(n):
        w.append(0)
    for i in range(k):
        w = w - nu * 2 * numpy.dot(numpy.transpose(x), numpy.dot(x, w) - y)/ len(x)
    return w
```

```python
def secondMethodError():
    ans = []
    for k in range(0, 2000, 100):
        ans.append(nrmse(ytest, numpy.dot(xtest, secondMethod(xtrain, ytrain, k, 1e-19))))
    return ans
```

```python
matplotlib.pyplot.plot(range(0, 2000, 100), secondMethodError())
matplotlib.pyplot.show()
```

# Генетический алгоритм

```python
def probGo (de, t):
    if (abs(de/t) > 100):
        return 100000
    return numpy.exp (-de/t)

def thirdMethod(x, y, k, nuTemp, initT):
    w = []
    for i in range(n):
        w.append(0)
    wNRMSE = nrmse (y, numpy.dot(x, w))
    ans = w
    ansNRMSE = wNRMSE
    t = initT;
    for i in range(k):
        helpw = []
        for j in range (n):
            helpw.append(w[j] + random.uniform(0.2, 0.3))
        helpwNRMSE = nrmse (y, numpy.dot(x, helpw))
        print (wNRMSE, helpwNRMSE)
        randVal = random.uniform(0, 1)
        a = 0
        if (randVal <= probGo(helpwNRMSE - wNRMSE, t)):
            a = 1
        if (helpwNRMSE < wNRMSE):
            a = 1
        if (a == 1):
            wNRMSE = helpwNRMSE
            w = helpw
        if (helpwNRMSE < ansNRMSE):
            ansNRMSE = helpwNRMSE
            ans = helpw
        t = initT * nuTemp / (i + 1)
    return ans
```

```python
def thirdMethodError():
    ans = []
    for k in range(0, 2000, 100):
        ans.append(nrmse(ytest, numpy.dot(xtest, thirdMethod(xtrain, ytrain, k, 0.08, 1))))
    return ans
```

```python
matplotlib.pyplot.plot(range(0, 2000, 100), thirdMethodError())
matplotlib.pyplot.show()
```
