```python
import numpy
import matplotlib.pyplot as graphs
import random
import math
```

```python
file = open("1.txt")
m = int(file.readline())
nTrain = int(file.readline())
xTrain = []
yTrain = []
for i in range (nTrain):
    tmp = file.readline().split(' ')
    xTrain.append([])
    for j in range (m):
        xTrain[i].append(int(tmp[j]))
    yTrain.append(int(tmp[m]))
    
nTest = int(file.readline())
xTest = []
yTest = []
for i in range (nTest):
    tmp = file.readline().split(' ')
    xTest.append([])
    for j in range (m):
        xTest[i].append(int(tmp[j]))
    yTest.append(int(tmp[m]))
```

# Calculate NRMSE - calculate RMSE as sqrt(summa((predicted - y)^2) / n). Then divide by ymax - ymin 

```python
#xTest and yTest here
def nrmse(F, tetta, y):
    predicted = F @ tetta
    summa = numpy.mean((predicted - y)**2)
    rmse = numpy.sqrt(summa)
    ymax = y[0]
    ymin = y[0]
    for i in range (len(y)):
        if (y[i] > ymax):
            ymax = y[i]
        if (y[i] < ymin):
            ymin = y[i]
    return rmse / (ymax - ymin)
```

# 1st method - Pseudoinverse matrix

```python
#answer is tetta*, minimize (F * tetta*) - y
def pseudoinverse():
    F = numpy.array(xTrain)
    u, s, vt = numpy.linalg.svd(F, full_matrices=False)
    Fplus = vt.T @ numpy.linalg.inv(numpy.diag(s)) @ u.T
    answerPseudo = Fplus @ yTrain
    return answerPseudo.T
    #F = numpy.array(xTrain)
    #Fplus = numpy.linalg.pinv(F)
    #answerPseudo = Fplus @ yTrain
    #return answerPseudo.T
```

# "NRMSE for Pseudoinverse matrix method"

```python
print (nrmse(numpy.array(xTest), pseudoinverse(), numpy.array(yTest)))
```

# 2st method - Gradient descent

```python
def gradientDescent(steps):
    w = numpy.zeros(m)
    mu = 1e-19
    for i in range(steps):
        predicted = numpy.array(xTrain) @ numpy.array(w)
        gradient = 2 * numpy.array(xTrain).T @ (predicted - numpy.array(yTrain))
        w = w * (1 - mu * 3) - mu * gradient
    return w
```

```python
results1 = []
for i in range (0, 5000, 500):
    results1.append(nrmse(numpy.array(xTest), gradientDescent(i), numpy.array(yTest)))
```

```python
graphs.plot(range(0, 5000, 500), results1)
graphs.xlabel('iterations')
graphs.ylabel('Lose function')
#graphs.hlines(nrmse(numpy.array(xTest), pseudoinverse(), numpy.array(yTest)), 0, 5000)
graphs.show()
```

```python
print (nrmse(numpy.array(xTest), gradientDescent(100), numpy.array(yTest)))
```

# 3rd method - Simulated anneal

```python
#optimise Energy - mse, T - temperature, F - function of change from one tetta to another
def anneal(iterations):
    scale = 0.1
    step = 0.05
    tetta = numpy.zeros(m)
    tettaBest = tetta
    temp = 1
    mse = nrmse(numpy.array(xTrain), tetta, numpy.array(yTrain))
    mseBest = mse
    for i in range (iterations):
        temp = temp * (1 - step)
        tettaNew = []
        for j in range (m):
            tettaNew.append(tetta[j] + scale * random.uniform(2, 3))
        mseNew = nrmse(numpy.array(xTrain), tettaNew, numpy.array(yTrain))
        if ((mseNew < mse) or (random.random() < math.exp(-1 * (mseNew - mse) / temp))):
            tetta = tettaNew
            mse = mseNew
            if (mseNew < mseBest):
                mseBest = mseNew
                tettaBest = tettaNew
    return tettaBest
```

```python
results2 = []
for i in range (0, 1000, 100):
    results2.append(nrmse(numpy.array(xTest), anneal(i), numpy.array(yTest)))
```

```python
graphs.plot(range(0, 1000, 100), results2)
graphs.xlabel('iterations')
graphs.ylabel('Lose function')
#graphs.hlines(nrmse(numpy.array(xTest), pseudoinverse(), numpy.array(yTest)), 0, 1000)
graphs.show()
```
