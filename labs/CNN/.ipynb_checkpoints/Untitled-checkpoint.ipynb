```python
import tensorflow.keras as keras
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import models
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score
```

# Ищем лучшую модель на датасете с цифрами

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train = x_train.reshape((60000, 28, 28, 1))
x_train = x_train.astype('float32') / 255
x_test = x_test.reshape((10000, 28, 28, 1))
x_test = x_test.astype('float32') / 255
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', # (3,3) - фильтр
                        input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)), # фильтр (2,2) для пулинга
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.Flatten(),
    layers.Dense(10, 'softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5, batch_size=64)
```

```python
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)
print('Функция потерь на тесте - ', test_loss, "\nТочность на тесте - ", test_acc)
```

# Теперь обучим и протестируем уже на модном датасете

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
save_test_images = test_images
save_test_labels = test_labels
```

```python
train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
model.fit(train_images, train_labels, epochs=10)
predictions = model.predict(test_images)
```

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('Функция потерь на тесте - ', test_loss, "\nТочность на тесте - ", test_acc)
```

# Время матриц

```python
def matrix(true, classified):
    conf_matrix = confusion_matrix(true, classified)
    display = ConfusionMatrixDisplay(conf_matrix, display_labels = class_names)
    fig = plt.figure(figsize=(12, 10))
    accuracy = accuracy_score(true, classified)
    display.plot(ax=fig.add_subplot(111), values_format="d")
    plt.show()
```

```python
ans_labels = []
for i in predictions:
    ans_labels.append(np.argmax(i))
true_labels = []
for i in test_labels:
    true_labels.append(np.argmax(i))
matrix(np.array(true_labels), np.array(ans_labels))
```

```python
anotherCF = []
for i in range(len(class_names)):
    anotherCF.append([])
    for j in range(len(class_names)):
        indMax = 0
        probMax = 0
        for k in range(len(predictions)):
            if true_labels[k] != i:
                continue
            if (predictions[k][j] >= probMax):
                indMax = k
                probMax = predictions[k][j]
        anotherCF[i].append(indMax)
    
```

```python
plt.figure(figsize=(17, 17))
for i in range(100):
    j = i % 10
    curImg = anotherCF[i // 10][j]
    plt.subplot(10,10,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(save_test_images[curImg], cmap=plt.cm.binary)
    plt.ylabel(class_names[save_test_labels[curImg]])
    plt.xlabel(class_names[j])
plt.show()
```
